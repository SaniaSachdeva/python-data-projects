```python
performance.dropna(inplace=True)
```

```python

performance.replace("Male", 0, inplace=True)
performance.replace("Female", 1, inplace=True)

performance.replace("Yes", 1, inplace=True)
performance.replace("No", 0, inplace=True)

performance.replace("Low", 0, inplace=True)
performance.replace("Medium", 1, inplace=True)
performance.replace("High", 2, inplace=True)

performance.replace("Public", 0, inplace=True)
performance.replace("Private", 1, inplace=True)

performance.replace("Negative", 0, inplace=True)
performance.replace("Neutral", 1, inplace=True)
performance.replace("Positive", 2, inplace=True)

performance.replace("High School", 0, inplace=True)
performance.replace("College", 1, inplace=True)
performance.replace("Postgraduate", 2, inplace=True)

performance.replace("Near", 0, inplace=True)
performance.replace("Moderate", 1, inplace=True)
performance.replace("Far", 2, inplace=True)



X, y= performance.drop('Exam_Score', axis=1), performance['Exam_Score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=87)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict( X_test)
lm.score(X_test, y_test)
```

```python
@app.get("/performance/sample")
def get_sample():
    return performance.drop('Exam_Score', axis=1).iloc[0].to_json(orient="index")
```

```python
@app.get("/performance/explain")
def get_expln():
    return """This web service predicts a student’s exam score using different parts of their academic life, study habits, and environment.

When you send data to the model, you include information such as how many hours the student studies, their attendance percentage, and their previous exam scores. Some fields use numbers instead of words — for example, parental involvement, access to resources, and motivation are encoded as 0 for low, 1 for medium, and 2 for high.

Other factors, such as extracurricular activities, internet access, and learning disabilities, are simple yes-or-no values represented as 0 or 1.

Habits like sleep hours, physical activity, and tutoring sessions are included as numeric values. Family background is represented through income level and parental education, while school-related aspects like teacher quality, school type, peer influence, and distance from home are also provided as numerical codes.

Gender is encoded as 0 for male and 1 for female.

When all these details are submitted, the model processes them together and returns a single number — the predicted exam score. This score represents how well the student is expected to perform based on the combined influence of their habits, environment, and personal factors."""
```

```python
@app.post("/performance/evaluate")
def post_evaluation():
    if request.is_json:
        person = request.get_json() #Get the JSON from postman.
        df_test = pd.DataFrame([person]) #Convert JSON to dataframe.
        result = lm.predict(df_test) #Test our result
        return '{}'.format(result[0]) #Return the result
    return {"error": "Request must be JSON"}
```

```python
                                                              
if __name__ == '__main__':
    app.run()
```
